```python
def busqueda(usuario, articulo, caracteristicas):
  return tabla[(tabla.Vendedor==str(usuario)) & (tabla.Artículo==str(articulo))].loc[:,list(caracteristicas)]

v=busqueda("Boris Stanford","Pepinillo",["Región", "Ganancia", "Unidades"])
print(v)
print(v.value_counts(ascending=True, dropna=True).plot.pie())
```
